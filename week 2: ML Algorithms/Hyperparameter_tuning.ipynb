```python
import pandas as pd
import numpy as np
from sklearn import svm
```

```python
from sklearn.datasets import load_iris
iris = load_iris()
dir(iris)
```

```python
iris.data
```

```python
input=pd.DataFrame(iris.data, columns=iris.feature_names)
input
```

```python
target=pd.DataFrame(iris.target)
target
```

```python
iris.target
```

```python
from sklearn import svm
```

```python
from sklearn.model_selection import GridSearchCV
gcv=GridSearchCV(svm.SVC(),{
    'C':[1,10,20,30],
    'kernel':['linear','poly','rbf','sigmoid']},
                 cv=5,return_train_score=False

)
gcv.fit(input,target)
gcv.cv_results_
```

```python
df1=pd.DataFrame(gcv.cv_results_)
```

```python
df1
```

```python
gcv.best_params_
```

```python
gcv.best_score_
```

```python
dir(gcv)
```

For reducing computation cost, we can use RandomSearchCV

```python
from sklearn.model_selection import RandomizedSearchCV
rcv=RandomizedSearchCV(svm.SVC(gamma='auto'),{
    'C':[1,10,20,30],
    'kernel':['linear','poly','rbf','sigmoid']},
                       cv=5,n_iter=10,return_train_score=False
)
```

```python
rcv.fit(iris.data,iris.target)
```

```python
df2=pd.DataFrame(rcv.cv_results_)
df2
```

```python
df2[['params','rank_test_score','mean_test_score']]
```

```python
rcv.best_score_
```
