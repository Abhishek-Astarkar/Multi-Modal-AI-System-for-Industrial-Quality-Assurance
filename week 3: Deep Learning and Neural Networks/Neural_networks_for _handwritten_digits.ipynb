```python
import tensorflow as tf
from tensorflow import keras
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
(x_train,y_train),(x_test,y_test)=keras.datasets.mnist.load_data()
```

scaling is important

```python
x_train=x_train/255
x_test=x_test/255
```

```python
x_test[8]
```

```python
x_train_flattened=x_train.reshape(len(x_train),28*28)
x_test_flattened=x_test.reshape(len(x_test),28*28)
```

For taking input data, need to flatten to 1D array

```python
x_test_flattened[1]
```

SImple model with no hidden layers

```python
model_simple=keras.Sequential([
    keras.layers.Dense(10,input_shape=(28*28,),activation='sigmoid'),
])
```

```python
model_simple.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
model_simple.fit(x_train_flattened,y_train,epochs=5)
```

```python
model_simple.evaluate(x_test_flattened,y_test)
```

```python
y_predicted=model_simple.predict(x_test_flattened)
y_predicted[9]
```

```python
np.argmax(y_predicted[9])
```

```python
plt.matshow(x_test[9])
```

```python
y_predicted
```

To directly return output, np.argmax finds max value and returns its index

```python
y_predicted_labels=[np.argmax(i) for i in y_predicted]
```

```python
y_predicted_labels[56]
```

```python
y_test[56]
```

Confusion matrix for analysis

```python
cm=tf.math.confusion_matrix(y_test,y_predicted_labels)
cm
```

confusion matrix for analysis

```python
import seaborn as sn
plt.figure(figsize=(10,7))
sn.heatmap(cm,annot=True,fmt="g")
plt.xlabel("Predicted label")
plt.ylabel("True label")
```

Model with hidden layers

```python
model=keras.Sequential([
    keras.layers.Dense(100,input_shape=(28*28,),activation='sigmoid'),
    keras.layers.Dense(10,activation='sigmoid'),
])
tb_callback=tf.keras.callbacks.TensorBoard(log_dir='logs/',histogram_freq=1)
model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy'],
)
```

```python
model.fit(x_train_flattened,y_train,epochs=10,callbacks=[tb_callback])
```

```python
y_pred=model.predict(x_test_flattened)
```

```python
y_pred_lbl=[np.argmax(j) for j in y_pred]
error=[]
for i in range(1000):

    error.append(y_pred_lbl[i]-y_test[i])

print(error)
```

```python
num=0
for i in range(1000):
    if error[i]!=0:
        num=num+1
print(num)
```

```python
conf_matrix=tf.math.confusion_matrix(y_test,y_predicted_labels)
print(conf_matrix)
```

```python
import seaborn as sn

plt.figure()
sn.heatmap(conf_matrix, annot=True, fmt="g")
plt.xlabel("Predicted label")
plt.ylabel("True label")
```
