```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('data/Melbourne_housing_FULL.csv')
df.head()
```

```python
df.columns
df1=pd.DataFrame(df)
df1.head()
```

```python
df.columns
```

```python
df.nunique()
```

```python
df1.isna().sum()
```

```python
df1=df[['Suburb', 'Rooms', 'Type', 'Method', 'SellerG', 'Regionname', 'Propertycount',
               'Distance', 'CouncilArea', 'Bedroom2', 'Bathroom', 'Car', 'Landsize', 'BuildingArea', 'Price']]
df1
```

```python
df1.isna().sum()
```

```python
df1[['Bathroom','Bedroom2','Car','Distance']]=df1[['Bathroom','Bedroom2','Car','Distance']].fillna(0)
df1.isna().sum()
```

```python
df1['Landsize']=df1['Landsize'].fillna(df1.Landsize.mean())
df1['BuildingArea']=df1['BuildingArea'].fillna(df1.BuildingArea.mean())
```

```python
df1.isna().sum()
```

```python
df1['Price']=df1['Price'].dropna()
```

```python
df1.dropna(inplace=True)
df1.isna().sum()
```

```python
df1.shape
```

```python
input=df1.drop(['Price'],axis='columns')
```

```python
target=df1['Price']
```

```python
input.head()
```

```python
input=pd.get_dummies(input,drop_first=True)
input.head()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(input,target,test_size=0.3,random_state=2)
```

Normal Linear Regression model

```python
from sklearn.linear_model import LinearRegression
regressor=LinearRegression()
regressor.fit(x_train,y_train)
```

```python
regressor.score(x_test,y_test)
```

```python
regressor.score(x_train,y_train)
```

```python
len(x_train)
```

```python
len(x_test)
```

L1(Lasso) and L2(Ridge) regularization

```python
from sklearn import linear_model
lasso=linear_model.Lasso(alpha=100,max_iter=1000,tol=0.1)
lasso.fit(x_train,y_train)
```

```python
lasso.score(x_test,y_test)
```

```python
lasso.score(x_train,y_train)
```

```python
ridge=linear_model.Ridge(alpha=100,max_iter=1000,tol=0.1)
ridge.fit(x_train,y_train)
```

```python
ridge.score(x_test,y_test)
```

```python
ridge.score(x_train,y_train)
```
