```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('data/heart.csv')
df
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
```

```python
df.Sex=le.fit_transform(df.Sex)
df.ChestPainType=le.fit_transform(df.ChestPainType)
df.RestingECG=le.fit_transform(df.RestingECG)
df.ExerciseAngina=le.fit_transform(df.ExerciseAngina)
df.ST_Slope=le.fit_transform(df.ST_Slope)
df
```

```python
input=df.drop('HeartDisease',axis='columns')
```

```python
input
```

```python
input.isna().sum()
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
input_scaled=sc.fit_transform(input)
input_scaled
```

```python
target=df['HeartDisease']
target
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(input_scaled,target,test_size=0.2,random_state=40)
```

```python
from sklearn.svm import SVC
svm=SVC()
```

```python
from sklearn.model_selection import cross_val_score
svm_raw=cross_val_score(svm,x_train,y_train,cv=5)
```

```python
svm_raw.mean()
```

Bagging

```python
from sklearn.ensemble import BaggingClassifier
```

```python
bag=BaggingClassifier(
  estimator=SVC(),
    n_estimators=100,
    max_samples=0.8,
    oob_score=True,
    random_state=0
)
```

```python
bag.fit(x_train,y_train)
```

```python
svm_bagged=cross_val_score(bag,x_train,y_train,cv=5)
```

```python
svm_bagged.mean()
```

```python
bag.oob_score_
```

try the same with Decision tree

```python
from sklearn.tree import DecisionTreeClassifier
tree=DecisionTreeClassifier()
tree.fit(x_train,y_train)
```

```python
cvs=cross_val_score(tree,x_train,y_train,cv=5)
cvs.mean()
```

```python
bag_dt=BaggingClassifier(
  estimator=DecisionTreeClassifier(),
    n_estimators=100,
    max_samples=0.8,
    oob_score=True,
)
bag_dt.fit(x_train,y_train)
```

```python
bag_dt.oob_score_
```
